### VIẾT CHƯƠNG TRÌNH SPLOM vẽ dữ liệu iris (4 chiều) Scatterplot Matrix

```python
import seaborn as sns
```

```python
def SPLOM(x):
    n,p=x.shape
    fig,ax=plt.subplots(p,p,figsize=(10,10))
    for i in range(p):
        for j in range(p):
            if i==j :
                sns.histplot(
                    x=x[:,i],
                    ax=ax[i,j]
                )
            if i!=j:
                ax[i,j].scatter(x[:,i],x[:,j],s=1)
    plt.tight_layout()
```

```python
SPLOM(iris.data)
```

```python
def Tinh_Hetoado(p):
    v=np.zeros((p,2))
    for i in range(p):
        v[i,0]=np.cos(2*np.pi/p*i)
        v[i,1]=np.sin(2*np.pi/p*i)
    return v 
```

```python
x_min=x.min(axis=0)
x_min
```

```python
x_scale=(x-x_min)/(x_max-x_min)
```

```python
n,p=x.shape
```

```python
t=np.linspace(0,2*np.pi,101)
x1=np.cos(t)
y1=np.sin(t)

S=np.zeros((p,2))
for i in range(p):
    S[i,0]=np.cos(np.pi*2*i/p)
    S[i,1]=np.sin(np.pi*2*i/p)
    
y=np.zeros((n,2))
for i in range(n):
    x_sum=x[i,:].sum()
    y_temp=np.dot(x[i,:],S)
    y[i,:]=y_temp/x_sum

plt.figure(figsize=(5,5))
plt.plot(x1,y1)
plt.scatter(y[:,0],y[:,1],c='blue',s=1)
plt.scatter(S[:,0],S[:,1],c='red',s=10)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def parallel_coordinates_plot(df, class_column=None, alpha=0.5, cmap="Set1"):

    df_normalized = (df - df.min()) / (df.max() - df.min())

    if class_column:
        unique_classes = df[class_column].unique()
        colors = sns.color_palette(cmap, len(unique_classes))
        class_map = {cls: colors[i] for i, cls in enumerate(unique_classes)}
        
        for cls in unique_classes:
            subset = df_normalized[df[class_column] == cls]
            plt.plot(subset.drop(columns=[class_column]).T, color=class_map[cls], alpha=alpha, label=str(cls))


    else:
        plt.plot(df_normalized.T, color="gray", alpha=alpha)

    plt.xticks(range(len(df.columns) - (1 if class_column else 0)), df.columns[:-1] if class_column else df.columns, rotation=20)
    plt.xlabel("Features")
    plt.ylabel("Normalized Value")
    plt.title("Parallel Coordinates Plot")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.datasets import load_iris

# Load dữ liệu Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target  

parallel_coordinates_plot(df, class_column="species")
```
